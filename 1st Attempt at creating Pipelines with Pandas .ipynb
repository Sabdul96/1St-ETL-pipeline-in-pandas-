```python
def mean_Bedrooms_by_group(dataframe, col): 
    #groups the data by a column and returns the means of the price per group 
    return dataframe.groupby(col).mean()
```

```python
## create a function 

def lowercase_column_name(dataframe): 
    #lower case all of the column headers 
    dataframe.columns = dataframe.columns.str.lower()
    #return dataframe 
    return dataframe 
```

```python
# create a pipeline that applies both functions 
(df.pipe(mean_Bedrooms_by_group, col ='Bedrooms')
    #then apply the lowercase function 
    .pipe(lowercase_column_name)
)
```
